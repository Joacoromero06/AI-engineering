‚ú®**FIGURES & AXES**

```python
import matplotlib.pyplot as plt
fig,(ax1,ax2)=plt.subplots(2,1,figsize=(7,5))#asign axes to fig, parameter set the division of the all figure
ax1.plot([1,2,3],[3,3,3]);ax1.set_title("plot 1")#(1,3)(2,3)(3,3)
ax2.plot([10,100,1000],[1,2,3]);ax2.set_title("plot 2")#(10,1)(100,2)(1000,3)
plt.show()
```

üçñ**MULTIPLE PLOTS IN ONE AXE**

```python
import matplotlib.pyplot as plt
fig,ax=plt.subplots(1,1,figsize=(7,5))
x=[1,2,3,4];y1=[1,4,9,16];y2=[1,2,3,4];y3=[-1,-2,-3,-4]
ax.plot(x,y1)
ax.plot(x,y2)
ax.plot(x,y3)
```

üèü**MULTIPLE TYPE OF PLOT IN ONE AXE**

```python
import matplotlib.pyplot as plt
import numpy as np

fig,axe=plt.subplots()
x=np.linspace(0,10,100)#creates a list with [0,...,10] where we have 100 points well spaced each other
y=np.sin(x)
axe.plot(x,y)
plt.show()
```
